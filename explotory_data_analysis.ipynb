## 4. Univariate Analysis

First lets separate our categorical and numerical columns

```python
numerical_cols = df.columns[:10]
categorical_cols = df.columns[10:]
print("Numerical columns:\n", numerical_cols)
print("\nCategorical columns:\n", categorical_cols)
```

#### (4-A) Histogram and box plots for numerical columns

```python
for col in numerical_cols:
	fig, axes = plt.subplots(1, 2, figsize=(25, 6))
	fig.suptitle(col+" (Histogram and Boxplot)", fontsize=20)

	skew_value = round(df[col].skew(), 2)
	axes[0].text(0.5, 0.95, f'Skewness: {skew_value}', ha='center', va='top', fontsize=12, transform=axes[0].transAxes)

	# Histogram Plot
	axes = axes.flatten()
	df[col].hist(grid=False,ax=axes[0])
	axes[0].set_ylabel("Count")
	axes[0].set_xlabel(col)

	# Box plot	
	sns.boxplot(x=df[col],ax=axes[1])

```

We see a lot of skewness and in the numerical columns. This is seen in histogram plot
#### (4-B) Now  doing a count plot for our categorical variables

```python
fig, axes = plt.subplots(4, 2, figsize = (18, 18))
fig.suptitle('Bar plot for all categorical variables in the dataset')
axes = axes.flatten()

for i, col in enumerate(categorical_cols):
	sns.countplot(ax = axes[i], x = col, data = df, color = 'blue', order = df[col].value_counts().index)
```

## 5. Data Transformation

Based on the univariate analysis, we have noticed that almost all the numerical columns are highly skewed. So we can do a log transformation to fix this

```python
# Apply log1p transformation to all numerical columns
for col in numerical_cols:
    df[col] = np.log1p(df[col])  # log1p is equivalent to log(x + 1)
```

**PageValue Relationships:** 
- [BounceRates] We see a high page value means a low bounce rate (user is engaged and moving through page)
	- where as (low  - middle) corresponds a high  bounce rates
	- As Page value goes beyond 2.5, the bounce rate decreases gradually meaning user is engaged
- [ExitRates] (low - middle) page values, resulted a **high exit rate** means (Page was often the last viewed in session) or user left early
	- (middle -  high) pagevalues, resulted a **low exit rate** means Page was a step in the journey; user kept browsing.

*To summarize: High Page values resuled in low bounce and exit rate and vice Versa*

Each feature belonging to ["Adminitrative", "Adminitrative_duration", "Informational", "Informational_duration", "ProductRelated", "ProductRelated_duration"] has a relationships with ["BounceRates", "ExitRates", "PageValues"] <br>

### (6-C) Lets explore the relationship between categorical values and numerical values


Lets do a barplot to show the relationship between a continious variable and categorical values  <br>

```python
def bar_plot_continious_features_against_a_category(continuous_cols, category):
	fig, axes = plt.subplots(2, 5, figsize=(25, 5)) 
	fig.suptitle(f"{category} vs \n[{', '.join(continuous_cols)}]")
	axes = axes.flatten()

	for i, num_col in enumerate(continuous_cols):
		df.groupby(category)[num_col].mean().sort_values(ascending=False).plot.bar(ax = axes[i])
		axes[i].set_ylabel(num_col)
```

```python
for cat_col in categorical_cols:
	bar_plot_continious_features_against_a_category(numerical_cols, cat_col)
```

#### Observations

*  A true Revenue value resulted in: 
	- High Average page value, low bounce and exit rates
	- High Average for each of the durations and their repective columns
* Weekend, Month and Region did not have much of an effect on the numerical columns
* New and Returning visitor seemed to spend more time
* Browser 13 has the highest page value
* Some traffic types and Operating System correlated to a higher values in the numerical columns

## 7. Multivariate Analysis

Lets first encode the data  using sklearn label encoding

```python
from sklearn.preprocessing import LabelEncoder

df_encoded = df.copy()

# Label encode categorical columns
le = LabelEncoder()
df_encoded['Month'] = le.fit_transform(df_encoded['Month'])
df_encoded['VisitorType'] = le.fit_transform(df_encoded['VisitorType'])
df_encoded['Weekend'] = df_encoded['Weekend'].astype(int)  # Convert boolean to 0/1
df_encoded['Revenue'] = df_encoded['Revenue'].astype(int)

df_encoded.describe().T
```

```python
df_encoded.sample(5)
```

```python

corr_matrix = df_encoded.corr()

plt.figure(figsize=(18, 8))
sns.heatmap(corr_matrix, annot=True, vmin=-1, vmax=1, cmap="coolwarm")
```

#### Observations
- All numerical columns have some positive and negative relationships with each other.
- Some Categorical features had relationships with numerical and categorical columns while others didnt.
